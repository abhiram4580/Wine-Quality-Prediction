# White wine Quality Classification

```python


import numpy as np 
import pandas as pd 
from matplotlib import pyplot as plt

```

```python

df = pd.read_csv('winequality-white.csv',sep=';')
print(df.shape)
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python

print('Null vals:',df.isnull().sum().sum())
df.info()
```

```python
df['quality label'] = df['quality'].apply(lambda x: 'low' if x<=5 else 'medium' if x<=7 else 'high')
print(df[['quality','quality label']].value_counts().sort_index())
df.head()
```

## EDA

### Descriptive Statistics

```python
round(df.describe(),2)
```

## Preprocessing

```python


from sklearn.preprocessing import StandardScaler,LabelEncoder

encoder = LabelEncoder()
df['quality label'] = encoder.fit_transform(df['quality label'])

X = df.drop('quality',axis=1)
y = df.quality

scaler = StandardScaler()
X = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
X.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=42)
print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

## Modeling

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import r2_score,mean_squared_error

model_rfc = RandomForestClassifier()
model_rfc.fit(X_train,y_train)
y_pred = model_rfc.predict(X_test)

print('R2:',r2_score(y_test,y_pred))
```
